Undergraduates average = 20

```python
# Load the data
race_ethnicity_df = pd.read_excel("um_race_ethnicity.xlsx")
enrollment_df = pd.read_excel("umich_enrollment_number.xlsx")
programs_schools_df = pd.read_excel("umich_programs_schools.xlsx")
schools_count_df = pd.read_excel("umich_schools_count.xlsx")

# Extract relevant distributions
age_distribution = {
    "Under 22": 0.91,
    "22 - 23": 0.07,
    "24 - 25": 0.01,
    "26 - 30": 0.005,
    "Over 30": 0.005,
}

gender_distribution = race_ethnicity_df.groupby("Gender").sum().iloc[:, -1]
gender_distribution /= gender_distribution.sum()

race_distribution = race_ethnicity_df.groupby("Category").sum().iloc[:, -1]
race_distribution /= race_distribution.sum()

program_distribution = programs_schools_df["Program"].value_counts(normalize=True)

# Function to generate specific racial combinations for "Two or More Races (URM)"
def generate_urm_race():
    urm_races = ["African American", "Hispanic American", "Native American", "Native Hawaiian/Pacific Islander"]
    return " & ".join(np.random.choice(urm_races, 2, replace=False))

# Generate 80 undergraduate profiles
np.random.seed(42)
profiles = []
for _ in range(80):
    gender = np.random.choice(gender_distribution.index, p=gender_distribution.values)
    race = np.random.choice(race_distribution.index, p=race_distribution.values)
    program = np.random.choice(program_distribution.index, p=program_distribution.values)
    age_group = np.random.choice(list(age_distribution.keys()), p=list(age_distribution.values()))
    
    # Assign age based on age group
    if age_group == "Under 22":
        age = np.random.normal(loc=18.5, scale=1)
    elif age_group == "22 - 23":
        age = np.random.normal(loc=22.5, scale=0.5)
    elif age_group == "24 - 25":
        age = np.random.normal(loc=24.5, scale=0.5)
    elif age_group == "26 - 30":
        age = np.random.normal(loc=28, scale=1)
    else:
        age = np.random.normal(loc=32, scale=2)
    
    age = round(age)
    
    # Assign specific racial combinations for "Two or More Races (URM)"
    if race == "Two or More Races (URM)":
        race = generate_urm_race()
    
    # Retrieve school based on program
    school = programs_schools_df.loc[programs_schools_df["Program"] == program, "School/College"].values
    school = school[0] if len(school) > 0 else "Unknown"
    
    profiles.append((age, gender, race, program, school))

# Create DataFrame for generated profiles
df_profiles = pd.DataFrame(profiles, columns=["age", "gender", "race", "program", "school"])
df_profiles["Age"] = df_profiles["Age"].round().astype(int)
```

```python
df_profiles.to_excel("umich_undergraduate_profiles.xlsx", index=False)
```
